```python
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Assuming your merged DataFrame is named 'data' with columns 'Price_VIX' and 'Price_SPX'

# Calculate daily returns
data['VIX_ret'] = data['Price_VIX'].pct_change()
data['SPX_ret'] = data['Price_SPX'].pct_change()
data.dropna(inplace=True)

def objective(weights, cov_matrix):
    """Calculates the portfolio variance (to minimize)."""
    portfolio_variance = weights.T @ cov_matrix @ weights
    return portfolio_variance

def optimize_portfolio(returns):
    """Optimizes portfolio weights using Markowitz approach."""
    cov_matrix = returns.cov()
    num_assets = len(returns.columns)

    # Initial weights (equal allocation)
    initial_weights = np.array([1 / num_assets] * num_assets)

    # Define bounds for weights (between 0 and 1)
    bounds = [(0, 1) for _ in range(num_assets)]

    # Constraint: sum of weights must equal 1
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

    # Optimization using SLSQP solver
    result = minimize(
        objective,
        initial_weights,
        args=(cov_matrix,),
        method='SLSQP',
        bounds=bounds,
        constraints=constraints
    )

    return result.x

def rebalance_strategy_1(data, spx_ret, lookback_period=3):
    """Rebalance based on SPX consecutive losses."""
    if (spx_ret[-lookback_period:].min() < -0.02) or (np.all(spx_ret[-lookback_period:] < 0)):
        return [0.2, 0.8]  # Increase VIX weight
    else:
        return [0.8, 0.2]  # Increase SPX weight

def rebalance_strategy_2(data, spx_ret, lookback_period=5, volatility_threshold=0.015):
    """Rebalance based on SPX volatility."""
    spx_volatility = data['SPX_ret'][-lookback_period:].std()
    if spx_volatility > volatility_threshold:
        return [0.3, 0.7]  # Increase VIX weight if volatility is high
    else:
        return [0.7, 0.3]  # Increase SPX weight if volatility is low

# Initialize lists to store portfolio returns and weights for each strategy
portfolio_returns = {'Strategy_1': [], 'Strategy_2': [], 'SPX': []}
weights_history = {'Strategy_1': [], 'Strategy_2': []}

# Backtesting loop
for i in range(252, len(data)):  # Start after the initial lookback period
    # Select returns for the past year
    returns = data[['VIX_ret', 'SPX_ret']][i-252:i]

    # Get current SPX return
    spx_ret = data['SPX_ret'][i-252:i]

    # --- Strategy 1: Consecutive Losses ---
    weights_1 = rebalance_strategy_1(data, spx_ret)
    weights_history['Strategy_1'].append(weights_1)
    portfolio_ret_1 = weights_1 @ data[['VIX_ret', 'SPX_ret']].iloc[i]
    portfolio_returns['Strategy_1'].append(portfolio_ret_1)

    # --- Strategy 2: Volatility Based ---
    weights_2 = rebalance_strategy_2(data, spx_ret)
    weights_history['Strategy_2'].append(weights_2)
    portfolio_ret_2 = weights_2 @ data[['VIX_ret', 'SPX_ret']].iloc[i]
    portfolio_returns['Strategy_2'].append(portfolio_ret_2)

    # --- SPX Benchmark ---
    portfolio_returns['SPX'].append(data['SPX_ret'].iloc[i])

# Create DataFrame for analysis
results = pd.DataFrame(portfolio_returns, index=data.index[252:])

# Calculate cumulative returns
results['Cumulative_ret_1'] = (1 + results['Strategy_1']).cumprod()
results['Cumulative_ret_2'] = (1 + results['Strategy_2']).cumprod()
results['Cumulative_ret_SPX'] = (1 + results['SPX']).cumprod()

# Plot cumulative returns
results[['Cumulative_ret_1', 'Cumulative_ret_2', 'Cumulative_ret_SPX']].plot()
plt.title('Backtesting Results - Cumulative Returns')
plt.legend(['Strategy 1', 'Strategy 2', 'SPX'])
plt.show()
```

```python
########## FINDING ALPHA ##############
#########

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def test_predictors(alltick_norm, predictors_df):
    """
    Tests the predictive power of columns in predictors_df for next month's stock returns 
    in alltick_norm.

    Args:
        alltick_norm: DataFrame of stock prices, with tickers as columns and dates as rows.
        predictors_df: DataFrame of potential predictors, with dates as rows.

    Returns:
        DataFrame: Results of regressions for each ticker, including R-squared and coefficients.
    """

    # Calculate next month's returns for each stock
    returns_df = alltick_norm.pct_change(periods=1).shift(-1)

    results = []
    for ticker in alltick_norm.columns:
        # Prepare data for regression
        X = predictors_df.loc[returns_df.index]  # Align predictors with returns
        y = returns_df[ticker].dropna()

        # Check for sufficient data points for regression
        if len(y) < len(X.columns) + 1:
            print(f"Insufficient data for regression for ticker: {ticker}")
            continue  # Skip regression if not enough data

        # Fit linear regression model
        model = LinearRegression()
        model.fit(X, y)

        # Evaluate model performance (R-squared)
        r2 = r2_score(y, model.predict(X))

        # Store results
        results.append({
            'Ticker': ticker,
            'R-squared': r2,
            **{predictor: coef for predictor, coef in zip(X.columns, model.coef_)}
        })

    return pd.DataFrame(results)



predictors_df = 
regression_results = test_predictors(alltick_norm, predictors_df)
print(regression_results)
```

```python
def calculate_conviction(qty, delta_own, scaler = 100):
    """Calculates a conviction score based on Qty and Î”Own."""

    # Normalize Qty (e.g., divide by average Qty in your dataset)
    normalized_qty = qty / 100000  # Example normalization, adjust as needed

    # Convert Î”Own to a numerical value (handle cases like 'New', '>999%')
    if delta_own == 'New':
        delta_own_num = 4.0  # High conviction for new positions
    elif delta_own == '>999%':
        delta_own_num = 4.0  # Very high conviction for large increases
    else: delta_own_num = delta_own

    # Combine normalized Qty and delta_own_num into a conviction score
    # (You can adjust the weights and formula as needed)
    conviction_score = 4 * normalized_qty + scaler* delta_own_num

    return conviction_score/20
```

```python
inside2 = inside.copy()

def merge_rows_by_ticker_date(inside):
    """
    Merges rows in the "inside" DataFrame where the same ticker appears multiple times
    on the same date, according to the specified aggregation rules.

    Args:
        inside: DataFrame containing ticker, date, qty, Delta own, Price, and other columns.

    Returns:
        DataFrame: Merged DataFrame with aggregated rows.
    """

    # Group by ticker and date
    grouped = inside.groupby(['Ticker', 'FilingDate'])

    # Define aggregation functions for each column
    agg_funcs = {
        'Qty': 'sum',
        'Î”Own': 'mean',
        'Price': 'mean'
    }

    # Get the first value for all other columns
    first_vals = {col: 'first' for col in inside.columns if col not in agg_funcs}

    # Combine aggregation functions
    agg_funcs.update(first_vals)

    # Apply aggregation and reset index
    merged_inside = grouped.agg(agg_funcs).reset_index()

    return merged_inside

# Example usage:
# Assuming your DataFrame is called "inside"

merged_inside = merge_rows_by_ticker_date(inside2)
print(merged_inside)
```

```python
# The prediction model implemented in the article is called BAROW (Batch-update AROW for Regression). It's an extension of the AROW (Adaptive Regularization of Weights) algorithm, specifically designed for online learning in non-stationary environments like financial markets.



class BAROW:
    def __init__(self, r, d):
        """
        Initialize BAROW with hyperparameter r and feature dimension d.
        """
        self.r = r
        self.d = d
        self.mu = np.zeros(d)
        self.Sigma = np.eye(d)

    def update(self, X, Y):
        """
        Update the model parameters (mu and Sigma) based on a batch of 
        features (X) and corresponding returns (Y).
        
        Args:
            X: A numpy array of shape (K, d) where K is the batch size 
               and d is the feature dimension.
            Y: A numpy array of shape (K, 1) containing the returns.
        """
        R = self.r * X.shape[0] # As defined in the article
        
        # Calculate the updated Sigma using equation (2a)
        self.Sigma = self.Sigma - self.Sigma @ X.T @ np.linalg.inv(R * np.eye(X.shape[0]) + X @ self.Sigma @ X.T) @ X @ self.Sigma

        # Calculate the updated mu using equation (2b)
        self.mu = self.mu - self.Sigma @ X.T @ np.linalg.inv(R * np.eye(X.shape[0]) + X @ self.Sigma @ X.T) @ (X @ self.mu - Y)
        
    def predict(self, X):
        """
        Predict returns based on the current model parameters.
        
        Args:
            X: A numpy array of shape (K, d) where K is the number of 
               instances to predict and d is the feature dimension.
        
        Returns:
            A numpy array of shape (K, 1) containing the predicted returns.
        """
        return X @ self.mu
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import ortho_group

def prepare_data(df, train_tickers, test_tickers, D=250, train_period=754):
    """
    Prepares the training and testing datasets for the challenge.

    Args:
        df: DataFrame with stock data (dates as index, tickers as columns).
        train_tickers: List of tickers for training data.
        test_tickers: List of tickers for testing data.
        D: Time depth parameter (default: 250).
        train_period: Length of the training period (default: 754).

    Returns:
        X_train: Training data (returns).
        X_test: Testing data (returns).
    """

    # Calculate returns 
    returns = df.pct_change().dropna()

    # Extract training and testing data
    X_train = returns[train_tickers].iloc[-train_period:]
    X_test = returns[test_tickers].iloc[-train_period:] 

    return X_train, X_test 

# Example usage:
train_tickers = df.columns[:50] # First 50 tickers for training
test_tickers = df.columns[50:100] # Next 50 tickers for testing
X_train, X_test = prepare_data(df, train_tickers, test_tickers)
```

```python
def calculate_factors(returns, A, D):
    """
    Calculates factors (F_lt) based on past returns and matrix A.

    Args:
        returns: DataFrame of stock returns.
        A: Matrix with orthonormal columns (coefficients for past returns).
        D: Time depth parameter.

    Returns:
        factors: DataFrame of calculated factors.
    """

    factors = pd.DataFrame(index=returns.index)
    for l in range(A.shape[1]):
        factor = pd.Series(0, index=returns.index)
        for k in range(1, D + 1):
            factor += A[k-1, l] * returns.shift(k)
        factors[f'F{l+1}'] = factor.dropna()
    
    return factors

# Example usage (assuming A is already calculated):
factors_train = calculate_factors(X_train, A, D)
```

```python
def fit_beta(factors, returns):
    """
    Fits beta coefficients using linear regression.

    Args:
        factors: DataFrame of factors.
        returns: DataFrame of stock returns.

    Returns:
        beta: Fitted beta coefficients.
    """
    model = LinearRegression()
    model.fit(factors, returns)
    beta = model.coef_

    return beta

# Example usage:
beta = fit_beta(factors_train, X_train) 
```

```python
def calculate_metric(A, beta, returns_test, factors_test):
    """
    Calculates the metric as defined in the challenge description.

    Args:
        A: Matrix with orthonormal columns.
        beta: Fitted beta coefficients.
        returns_test: DataFrame of test set returns.
        factors_test: DataFrame of test set factors.

    Returns:
        metric: Calculated metric value.
    """

    S_predictions = factors_test @ beta.T

    orthonormal_check = np.allclose(A.T @ A, np.eye(A.shape[1]), atol=1e-10) 

    if orthonormal_check:
        metric = np.mean((returns_test * S_predictions).sum(axis=1) / np.sqrt((S_predictions**2).sum(axis=1)))
    else:
        metric = -1

    return metric

# Example usage (assuming factors_test are calculated using A and X_test):
metric = calculate_metric(A, beta, X_test, factors_test)
```

```python
def qrt_benchmark(X_train, D=250, F=10, N_iter=1000):
    """
    Implements the QRT benchmark described in the challenge.

    Args:
        X_train: Training data (returns).
        D: Time depth parameter (default: 250).
        F: Number of factors (default: 10).
        N_iter: Number of iterations (default: 1000).

    Returns:
        best_A: Best A matrix found.
        best_beta: Best beta coefficients found.
        best_metric: Best metric achieved.
    """

    best_metric = -float('inf')
    best_A = None
    best_beta = None

    for _ in range(N_iter):
        M = np.random.normal(0, 1, size=(D, D))
        A = ortho_group.rvs(dim=D)[:, :F] # Using ortho_group for orthonormal columns

        factors_train = calculate_factors(X_train, A, D)
        beta = fit_beta(factors_train, X_train)

        factors_test = calculate_factors(X_test, A, D)  # Calculate test set factors
        metric = calculate_metric(A, beta, X_test, factors_test)

        if metric > best_metric:
            best_metric = metric
            best_A = A
            best_beta = beta

    return best_A, best_beta, best_metric

# Example usage:
best_A, best_beta, best_metric = qrt_benchmark(X_train)
```

```python
####### LEARN A and beta 

import numpy as np
from pymanopt.manifolds import Stiefel
from pymanopt import Problem
from pymanopt.solvers import SteepestDescent, ConjugateGradient, TrustRegions


def cost_function(A, X_train, X_test, D, F, beta):
    """
    Cost function to be minimized on the Stiefel manifold.

    Args:
        A: Matrix on the Stiefel manifold (orthonormal columns).
        X_train: Training data (returns).
        X_test: Testing data (returns).
        D: Time depth parameter.
        F: Number of factors.
        beta: Beta coefficients.

    Returns:
        cost: The negative of the metric (to be minimized).
    """

    factors_train = calculate_factors(X_train, A, D)
    factors_test = calculate_factors(X_test, A, D)

    metric = calculate_metric(A, beta, X_test, factors_test)
    return -metric # Minimize the negative of the metric


def optimize_A_stiefel(X_train, X_test, D, F, beta_init, solver_type="SteepestDescent"):
    """
    Optimizes A on the Stiefel manifold using the specified solver.

    Args:
        X_train: Training data (returns).
        X_test: Testing data (returns).
        D: Time depth parameter.
        F: Number of factors.
        beta_init: Initial beta coefficients.
        solver_type: Type of solver to use (default: "SteepestDescent").
                            Options: "SteepestDescent", "ConjugateGradient", "TrustRegions"

    Returns:
        optimized_A: Optimized A matrix.
        optimized_beta: Optimized beta coefficients.
    """

    # Define the manifold
    manifold = Stiefel(D, F)

    # Define the cost function (using a lambda to pass other parameters)
    cost = lambda A: cost_function(A, X_train, X_test, D, F, beta_init)

    # Define the problem
    problem = Problem(manifold=manifold, cost=cost)

    # Choose the solver
    if solver_type == "SteepestDescent":
        solver = SteepestDescent()
    elif solver_type == "ConjugateGradient":
        solver = ConjugateGradient()
    elif solver_type == "TrustRegions":
        solver = TrustRegions()
    else:
        raise ValueError("Invalid solver_type. Choose from 'SteepestDescent', 'ConjugateGradient', or 'TrustRegions'")

    # Solve the problem
    optimized_A = solver.solve(problem)

    # Re-fit beta with the optimized A
    factors_train = calculate_factors(X_train, optimized_A, D)
    optimized_beta = fit_beta(factors_train, X_train)

    return optimized_A, optimized_beta

# Example usage:
D = 250  # Time depth
F = 10   # Number of factors
beta_init = np.random.rand(X_train.shape[1], F) # Initialize beta randomly

# Optimize A using Riemannian Gradient Descent
optimized_A_steepest, optimized_beta_steepest = optimize_A_stiefel(X_train, X_test, D, F, beta_init, solver_type="SteepestDescent")

# Optimize A using Conjugate Gradient
optimized_A_conjugate, optimized_beta_conjugate = optimize_A_stiefel(X_train, X_test, D, F, beta_init, solver_type="ConjugateGradient")

# ... (You can similarly use TrustRegions solver) 
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.constraints import Constraint

class OrthonormalConstraint(Constraint):
    """
    Custom constraint to enforce orthonormality on weight matrix.
    """
    def __call__(self, w):
        # Use Gram-Schmidt orthogonalization
        q, r = tf.linalg.qr(w)
        return q * tf.sign(tf.linalg.diag_part(r))

# Define the autoencoder model
input_dim = X_train.shape[1]  # Number of stocks
latent_dim = F  # Number of factors (bottleneck layer size)

# Encoder
encoder_input = Input(shape=(input_dim,))
encoder_hidden = Dense(128, activation='relu')(encoder_input)
encoder_output = Dense(latent_dim, activation='linear', 
                       kernel_constraint=OrthonormalConstraint())(encoder_hidden) 

# Decoder
decoder_input = Input(shape=(latent_dim,))
decoder_hidden = Dense(128, activation='relu')(decoder_input)
decoder_output = Dense(input_dim, activation='linear')(decoder_hidden)

# Create the encoder and decoder models
encoder = Model(encoder_input, encoder_output, name='encoder')
decoder = Model(decoder_input, decoder_output, name='decoder')

# Create the autoencoder model
autoencoder = Model(encoder_input, decoder(encoder_output), name='autoencoder')

# Compile the model
autoencoder.compile(optimizer='adam', loss='mse')

# Train the model
autoencoder.fit(X_train, X_train, epochs=100, batch_size=32) # Adjust parameters as needed

# Extract A (orthonormal weights of the bottleneck layer)
A = encoder.layers[-1].get_weights()[0]

# Fit beta (using linear regression on the learned factors)
factors_train = encoder.predict(X_train)
beta = fit_beta(pd.DataFrame(factors_train), X_train)
```

```python
from tensorflow.keras.layers import Lambda, K
from tensorflow.keras import backend as K

# Define the encoder network
encoder_input = Input(shape=(input_dim,))
encoder_hidden = Dense(128, activation='relu')(encoder_input)
z_mean = Dense(latent_dim, name='z_mean')(encoder_hidden)
z_log_var = Dense(latent_dim, name='z_log_var')(encoder_hidden)

# Sampling function
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0., stddev=1.)
    return z_mean + K.exp(z_log_var / 2) * epsilon

z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])

# Define the decoder network
decoder_hidden = Dense(128, activation='relu')(z)
decoder_output = Dense(input_dim, activation='linear')(decoder_hidden)

# Define the VAE model
vae = Model(encoder_input, decoder_output)

# Define the VAE loss function (reconstruction loss + KL divergence)
reconstruction_loss = tf.keras.losses.mse(encoder_input, decoder_output)
kl_loss = 1 + z_log_var - K.square(z_mean) - K.exp(z_log_var)
kl_loss = K.sum(kl_loss, axis=-1)
kl_loss *= -0.5
vae_loss = K.mean(reconstruction_loss + kl_loss)

vae.add_loss(vae_loss)
vae.compile(optimizer='adam')

# Train the VAE model
vae.fit(X_train, epochs=100, batch_size=32)

# Extract A (weights connecting latent space to input)
A = vae.get_layer('z_mean').get_weights()[0]

# Orthonormalize A
A = ortho_group.rvs(dim=A.shape[1])[:, :F]  # Orthonormalize using ortho_group

# Fit beta (using linear regression on the encoded data)
factors_train = vae.get_layer('z_mean').output
encoder_model = Model(encoder_input, factors_train)
encoded_data = encoder_model.predict(X_train)
beta = fit_beta(pd.DataFrame(encoded_data), X_train)
```
